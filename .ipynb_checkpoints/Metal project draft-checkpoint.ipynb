How are bands distributed by country? 

```python
fig, ax = plt.subplots(figsize=[15, 5])
bands['Country'].value_counts().head(15).plot.bar(ax=ax)
ax.set_title("Metal bands by country", fontsize=BIG_FONT)
ax.set_ylabel("Number of bands", fontsize=MED_FONT)
ax.set_xlabel("Country", fontsize=MED_FONT)
plt.xticks(rotation=45);
```

This plot reminds me of [Zipf's law](https://en.wikipedia.org/wiki/Zipf%27s_law), an emperical law stating that the frequency of a value is inversely proportional to its rank-order. Take, for example, the populations of U.S. cities: 

| City | Population (in millions) |
| -- | -- |
| New York | 8.5 |
| Los Angeles | 3.9 | 
| Chicago | 2.7 |

The population of Los Angeles (in 2nd place) is approximately 1/2 that of New York, and the population of Chicago (in 3rd place) is roughly 1/3 that of New York. 

Do the number of metal bands in each country actually follow this distrubution? 

Generate a new DataFrame containing the rank and number of bands for each country and make a log-log plot. If this plot appears linear, then this distribution follows Zipf's law

```python
order = pd.DataFrame(
    {'rank'     : range(1, len(bands['Country'].value_counts())+1),
     'num_bands': bands['Country'].value_counts().values})

fig, ax = plt.subplots(figsize=[10, 5])
sns.scatterplot(data=order, x='rank', y='num_bands')
plt.xscale('log')
plt.yscale('log')
ax.set_title("Log rank vs. log of num_bands for each country", fontsize=BIG_FONT)
ax.set_xlabel("rank", fontsize=MED_FONT)
ax.set_ylabel("number of bands", fontsize=MED_FONT);
```

It looks like this plot was linear for the first 3 data points but quickly became non-linear. Evidently, the number of metal bands by country does NOT follow Zipf's law.

```python
folk_metal_bands = agg_bands(bands, genre_filter = "Folk")
folk_metal_bands.head()
```

```python
# Generate X, Y pairs
folk_metal_bands = GeoDataFrame(
    world.merge(agg_bands(bands, genre_filter = "Folk"), how="outer", right_on="Country", left_on="name")
)
folk_metal_bands = folk_metal_bands.drop(["gdp_md_est", "Country"], 1)
folk_metal_bands['num_bands'] = folk_metal_bands['num_bands'].fillna(0)
folk_metal_bands['bands per capita'] = folk_metal_bands.num_bands / (folk_metal_bands.pop_est / 1e5)
folk_metal_bands.head()
```

```python
pairs = forests.merge(folk_metal_bands, how="outer", left_on="Country Name", 
                      right_on="name")#[["Country Name", "2016", "num_bands"]
pairs = pairs.drop(pairs[pairs['name'].isna()].index)
pairs = pairs[["name", "2016", "bands per capita", "pop_est", "continent"]]
pairs = pairs.rename(columns = {"2016": "percent forest coverage"})
pairs.head()
```

```python
sns.scatterplot(x=pairs["percent forest coverage"].values, y=pairs["bands per capita"].values)
```

```python
pairs.sort_values(by="percent forest coverage", ascending=False).head()
```

- What about Europe only? 

```python
fig, ax = plt.subplots(figsize=[10, 6])

sns.scatterplot(x=pairs["percent forest coverage"][pairs['continent']=='Europe'].values, 
                y=pairs["bands per capita"][pairs['continent']=='Europe'].values, 
                ax=ax, 
                s=100)
BIG_FONT = 18
MED_FONT = 15
SMALL_FONT = 12

ax.set_title("% land area covered by forests vs. folk metal bands per capita\n", fontsize=BIG_FONT)
ax.set_ylabel("folk metal bands per 100,000", fontsize=MED_FONT)
ax.set_xlabel("% forest coverage", fontsize=MED_FONT);
```

There does appear to be a correlation between forest coverage and folk metal bands per capita

```python
data = pairs[pairs['continent']=='Europe']
data = data.fillna(0)
data.head(50)
```

```python


X = data.iloc[:, 1].values.reshape(-1, 1)
y = data.iloc[:, 2].values.reshape(-1, 1)

```

```python
import matplotlib.patches as mpatches
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

r_square = round(regressor.score(X_train, y_train), 4)
r_label = f"$R^2$ = {r_square}"

b = round(regressor.intercept_[0], 4)
m = round(regressor.coef_[0][0], 4)

fig, ax = plt.subplots(figsize=[10, 6])


fit_label = f"y = {m}x + {b}"

sns.lineplot(x=X_train.reshape(1,-1)[0], y=regressor.predict(X_train).reshape(1,-1)[0], color='b', label=fit_label)
sns.scatterplot(x=pairs["percent forest coverage"][pairs['continent']=='Europe'].values, 
                y=pairs["bands per capita"][pairs['continent']=='Europe'].values, 
                ax=ax, 
                s=100, 
                color='r', 
                label='data')


plt.plot([], [], ' ', label=r_label)
handles, labels = plt.gca().get_legend_handles_labels()
order = [2, 0, 1]

plt.legend([handles[idx] for idx in order],
           [labels[idx] for idx in order]
          )

ax.set_title("% land area covered by forests vs. folk metal bands per capita\n", fontsize=BIG_FONT)
ax.set_ylabel("folk metal bands per 100,000", fontsize=MED_FONT)
ax.set_xlabel("% forest coverage", fontsize=MED_FONT)
plt.savefig('metal.png')
plt.show()
```
